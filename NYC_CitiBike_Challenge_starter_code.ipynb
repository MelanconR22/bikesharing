```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
bike_data_df = pd.read_csv('201908-citibike-tripdata.csv')
bike_data_df.head()
```

```python
# 2. Check the datatypes of your columns. 
bike_data_df.info()
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
#bike_data_df['tripduration'] = pd.to_datetime(bike_data_df['tripduration']).dt.strftime('%D:%H:%M:%S')
#bike_data_df['tripduration'] = pd.to_datetime(bike_data_df['tripduration'].dt.total_seconds() / 60)

bike_data_df['tripduration'] = pd.to_datetime(bike_data_df['tripduration'], unit='s')
bike_data_df.head()
```

```python
# 4. Check the datatypes of your columns. 
bike_data_df.info()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
bike_data_df.to_csv('citibike_tripdata.csv', index=False)
```
